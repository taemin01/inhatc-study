```python
import numpy as np
import pandas as pd
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler

# 데이터 세트를 읽어들인다.
df = pd.read_csv("wineqrev.csv", sep=',')

X=df.iloc[:,0:11]
y=df.iloc[:,11]

### 0과 1사이로 정규화
mms=MinMaxScaler()
mms.fit(X)
print(X.head())
X_mms=mms.transform(X)
X = pd.DataFrame(X_mms, columns=X.columns, index=list(X.index.values))

print(X.head())
#########################################################################3
# code 작성
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.2, shuffle=True)

model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(20, activation='relu'))  
model.add(Dense(20, activation='relu'))
model.add(Dense(1))
model.summary()

# 여기서 metrics를 설정해야 밑 evaluate에서 지표에 대한 값을 반환해준다.
model.compile(optimizer='adam', metrics = ['mean_squared_error'], loss='mean_squared_error')

early_stopping = EarlyStopping(monitor='val_loss', patience=20)

modelPath = "wineqr_assignment.keras"

checkPoint = ModelCheckpoint(filepath=modelPath, monitor='val_loss', verbose=1, save_best_only=True)

model.fit(X_train, y_train, validation_split=0.25, epochs=2000, batch_size=20, callbacks=[early_stopping, checkPoint])

#############################################################################3
# 테스트 결과를 출력합니다.
score=model.evaluate(X_test, y_test)
# evaluate는 두 가지 값을 반환한다.
# 1. 손실 값(loss) : 첫 번째 값 즉 score[0]
# 2. 지표(metric) : 두 번째 값 score[1] 이것은 모델을 평가 할 때 설정한 지표(정확도, MSE)다.

print(model.predict(X_test).flatten())
print('model performance(mse):', score[1])
```
